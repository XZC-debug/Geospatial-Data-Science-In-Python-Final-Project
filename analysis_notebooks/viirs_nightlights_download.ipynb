本notebook用于下载NASA Black Marble VIIRS夜间灯光数据

**数据产品:**
- VNP46A2: 每日夜间灯光数据
- VNP46A3: 月度夜间灯光数据

**数据用途:** 夜间亮度、光照强度变化分析

````python
# 导入必要的库
import earthaccess
import os
from datetime import datetime
import pandas as pd
````

## 2. 登录NASA Earthdata账号

**首次使用需要:**
1. 注册NASA Earthdata账号: https://urs.earthdata.nasa.gov/users/new
2. 运行下面的代码输入用户名和密码

登录信息会被保存，之后不需要重复输入

````python
# 登录NASA Earthdata
# 第一次运行会提示输入用户名和密码
auth = earthaccess.login()

# 如果已经保存了登录信息，会自动登录
# 如果想手动输入，使用: earthaccess.login(strategy="interactive")
````

## 3. 搜索数据

可以根据以下条件搜索数据:
- 时间范围
- 地理边界 (bounding box)
- 数据产品类型

````python
# 设置搜索参数
# 费城地区 2022-2024 三年的月度夜间灯光数据

# 选择数据产品：VNP46A3（月度数据）
short_name = "VNP46A3"  # 月度数据，适合长期趋势分析

# 设置时间范围：2022-2024 三年
start_date = "2022-01-01"
end_date = "2024-12-31"

# 设置地理范围：费城及周边地区
# 格式: (最小经度, 最小纬度, 最大经度, 最大纬度)
bounding_box = (-75.28, 39.87, -74.96, 40.14)  # 费城区域

print(f"搜索参数:")
print(f"数据产品: {short_name} (月度夜间灯光数据)")
print(f"时间范围: {start_date} 到 {end_date} (三年)")
print(f"地理范围: 费城及周边地区")
print(f"经度: {bounding_box[0]}° 到 {bounding_box[2]}°")
print(f"纬度: {bounding_box[1]}° 到 {bounding_box[3]}°")
````

````python
# 搜索数据
results = earthaccess.search_data(
    short_name=short_name,
    temporal=(start_date, end_date),
    bounding_box=bounding_box
)

print(f"\n找到 {len(results)} 个数据文件")

# 显示前5个结果的信息
if len(results) > 0:
    print("\n前5个文件:")
    for i, granule in enumerate(results[:5]):
        print(f"{i+1}. {granule['umm']['GranuleUR']}")
````

## 4. 下载数据

**注意事项:**
- VIIRS数据文件通常较大 (每个文件约100-500MB)
- 建议先下载少量数据测试
- 数据会保存到指定的文件夹中

````python
# 设置下载目录到你的finaldata文件夹
download_dir = r"D:\pen\MUSA5500GeospatialDataScienceInPython\final\finaldata"
os.makedirs(download_dir, exist_ok=True)

print(f"数据将保存到: {download_dir}")
````

````python
# 下载数据
# 这里先下载前3个文件作为测试
# 如果要下载全部，改为 results[:] 或 results

num_files_to_download = min(3, len(results))  # 先下载3个文件测试

print(f"开始下载 {num_files_to_download} 个文件...\n")

downloaded_files = earthaccess.download(
    results[:num_files_to_download],
    download_dir
)

print(f"\n下载完成! 共下载 {len(downloaded_files)} 个文件")
print(f"\n文件保存位置:")
for file in downloaded_files:
    print(f"  {file}")
````

## 5. 检查下载的文件

````python
# 列出下载目录中的所有文件
import glob

files = glob.glob(os.path.join(download_dir, "*.h5"))  # VIIRS数据是HDF5格式

print(f"下载目录中共有 {len(files)} 个HDF5文件:\n")
for file in files:
    size_mb = os.path.getsize(file) / (1024 * 1024)
    print(f"  {os.path.basename(file)} ({size_mb:.2f} MB)")
````

## 6. 批量下载全部数据（可选）

如果上面的测试成功，可以运行下面的代码下载所有搜索到的数据

## 常见问题解决

### 1. 登录失败
- 确保已注册NASA Earthdata账号
- 检查用户名和密码是否正确
- 尝试: `earthaccess.login(strategy="interactive")`

### 2. 搜索不到数据
- 检查时间范围是否正确
- 检查地理边界是否合理
- 尝试扩大搜索范围

### 3. 下载速度慢或中断
- NASA服务器在美国，国内下载可能较慢
- 可以分批下载，每次下载少量文件
- earthaccess会自动处理断点续传

### 4. 内存不足
- 不要一次下载太多文件
- 可以分批处理数据

### 5. 不同的搜索方式

```python
# 只按时间搜索（不限制地理范围）
results = earthaccess.search_data(
    short_name="VNP46A2",
    temporal=("2024-01-01", "2024-01-07")
)

# 搜索特定tile（如果知道tile编号）
results = earthaccess.search_data(
    short_name="VNP46A2",
    temporal=("2024-01-01", "2024-01-07"),
    granule_name="*h27v05*"  # 指定tile
)
```
